This file shows how to plot all the results of the experiments in the paper. Please use an appropriate df_path 

```python
df_path = 'Data/'

import pandas as pd 
```

```python
#plot time of binary choice 
import pandas as pd
round =100
accL = pd.read_pickle(df_path+'BINARY_CHOICE/libltimecase1.pkl')
accI = pd.read_pickle(df_path+'BINARY_CHOICE/ibltimecase1.pkl')
import matplotlib.pyplot as plt
plt.plot(range(round+1),accL.mean(axis =0)[:101],'o-', color='darkgreen', markersize=2, linewidth = 2,linestyle='-', label='SpeedyIBL')
plt.plot(range(round+1),accI.mean(axis =0)[:101],'o-', color='blue', markersize=2, linewidth = 2,linestyle='-', label='PyIBL')
plt.xlabel("Round")
plt.ylabel("Time (s)")
plt.title(f"Running time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plot performance of binary choice 

import pandas as pd
round =100
accL = pd.read_pickle(df_path+'BINARY_CHOICE/liblacccase1.pkl')
accI = pd.read_pickle(df_path+'BINARY_CHOICE/iblacccase1.pkl')
import matplotlib.pyplot as plt
plt.plot(range(round),accL.mean(axis =0)[:100],'o-', color='darkgreen', markersize=2, linewidth = 2, linestyle='-', label='SpeedyIBL')
plt.plot(range(round),accI.mean(axis =0)[:100],'o-', color='blue', markersize=2, linewidth = 2 ,linestyle='-', label='PyIBL')
plt.xlabel("Round")
plt.ylabel("PMax")
plt.title(f"Performance")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
